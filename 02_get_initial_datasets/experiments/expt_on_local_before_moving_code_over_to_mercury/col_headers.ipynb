```python
# Get categories using forward fill. For identifiers, there is no category and so we get NaN. 
# This formula allows us to get NaN + Annual Report Year -> Annual Report Year (because NaN + ' - ' = NaN) and 
# Characteristics of Roles + Individual Role -> Characteristics of Roles - Individual Role.
df.columns = (df.iloc[0,:].ffill() + ' - ').fillna('') + df.iloc[1,:]

# Remove the first two rows (which are column headers)
df = df.iloc[2:,:]
```
