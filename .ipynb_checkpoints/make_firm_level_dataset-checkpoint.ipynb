# Import Board Summary

```python
df = pd.read_csv(inputfilepath, low_memory=False)
df
```

```python
df1 = df.drop_duplicates(subset=['Annual Report Year', 'CompanyID*'], ignore_index=True)
df1
```

```python
df2 = df1.drop(['Director Type', 'DirectorID*', 'Individual Name', 'Individual Role'], axis=1)
# df2 = df1.iloc[:, 0:6]
df2
```

```python
# Replace NaNs with 0
df2 = df.replace('n.a', np.nan)
df2 = df2.fillna(0)
df2
```

```python
df2.columns
```

```python
df2.dtypes
```

```python
data_types_dict = {'Time in Role': 'float64', 'Time on Board': 'float64', 'Time in Company': 'float64', 
                   'Total Number of Quoted Boards to Date': 'int64', 'Age (Yrs)': 'int64', 'Number of Qualifications': 'int64',
                   'Bonus/ (Bonus&Salary)': 'float64', 'Bonus/ (Bonus&Salary)': 'float64', 'Equity Linked/ Total': 'float64',
                   'Performance/ Total': 'float64', 
                   'Total Directors on the Board': 'int64', 'Salary': 'float64', 'Total Annual Compensation': 'float64'}
df2 = df2.astype(data_types_dict)
df2.dtypes
```

# Groupby year and company

```python
agg_dict = {'Country':'first','ISIN':'first','Time in Role':'mean', 'Time on Board':'mean', 'Time in Company':'mean', 
            'Total Number of Quoted Boards to Date':'mean', 'Age (Yrs)':'mean', 'Number of Qualifications':'mean', 'Gender':list,
            'Nationality Mix':list,'Bonus/ (Bonus&Salary)':'mean', 'Equity Linked/ Total':'mean', 'Performance/ Total':'mean', 
            'Total Directors on the Board':'mean', 'Salary':'mean', 'Total Annual Compensation': 'mean', 
            'Region':'first','Mgmt_Type':'first'}
# CompanyID and CompanyID + Company Name both gave same number of rows
df2_groupby = df2.groupby(['Annual Report Year', 'CompanyID*', 'Company Name', 'Sector']).agg(agg_dict).reset_index()
df2_groupby
```

# Merge with Comapny Details

```python
df_companydetails = pd.read_csv(r"C:\Users\jasonjia\Dropbox\Projects\boardex_and_execucomp\output\filter_boardex_by_cols\output_from_mercury_combined_fragment_country_and_managementtype\Company Details_director_and_smde.csv", low_memory=False)
df_companydetails
```

```python
df_companydetails = df_companydetails[['CompanyID*', 'Bankers']]
df_companydetails
```

```python
df_merge = df2_groupby.merge(df_companydetails, how='left', on='CompanyID*')
df_merge
```

# Cleanup df and save final df into csv

```python
outputfilepath = Path(r"C:\Users\jasonjia\Dropbox\Projects\boardex_and_execucomp\output\boardex_firm_and_manager_level_dataset\boardex_firm_level_dataset.csv")
df_merge.to_csv(outputfilepath, index=False)
```
