```python
def get_df(filepath):
    row0 = pd.read_excel(filepath, header=None, nrows = 1)
    pct_nan = pd.isna(row0).sum().mean()
    if pct_nan > 0.5:
        print("% NaNs in first row:", pct_nan, "> 50%, taking header = row 1 by skipping row 0")
        df = pd.read_excel(filepath_sample, header=1)
    else:
        print("% NaNs in first row", pct_nan, "<= 50%, taking header = row 0 as usual")
        df = pd.read_excel(filepath_sample, header=0)
    
    return df
```

```python
def find_first_column_containing_companyid(df):
    for column in df.columns:
        if "companyid" in column.lower():
            return column
    return None
```

```python
companyid_list = pd.read_excel(companyid_listpath, header=None)[0].tolist()

for root, dirs, files in os.walk(datadir, topdown=False):
    for name in files:
        # Get filepath
        filepath = Path(os.path.join(root, name))
        print("Filepath:", filepath)
        # Open file correctly
        df = get_df(filepath)
        # Find first column header that contains 'companyid' in its name
        companyid = find_first_column_containing_companyid(df)
        
        if companyid == None:
            print("No column found that contains 'companyid', skipping df")
        else:
            # Create output filepath name
            inputfilepath = filepath_sample
            outputfilename = inputfilepath.stem + "_filtered" + inputfilepath.suffix
            outputfilepath = Path(inputfilepath.parent / outputfilename)
            # Filter df based on list of company ids provided
            df_filtered = df[df[companyid].isin(companyid_list)]
            # Save filtered df
            df_filtered.to_excel(outputfilepath, index=False)
            print("Saved filtered df to:", outputfilepath)
```
