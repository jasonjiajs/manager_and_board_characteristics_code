```python
import pandas as pd
from pathlib import Path
from rapidfuzz import process, fuzz
```

```python
# settings to display all columns and rows
pd.set_option("display.max_columns", None)
```

# Part 1: read excel

```python
# Get spreadsheet with isins that have 0 matches
df = pd.read_excel(r"C:\Users\jasonjia\Dropbox\Projects\boardex_and_execucomp\isins.xlsx", sheet_name='isin_no_match_even_with_matches')
```

```python
# Get spreadsheet with boardex company ids, ISINs and company names
boardex_df = pd.read_excel(r"C:\Users\jasonjia\Dropbox\Projects\boardex_and_execucomp\isins.xlsx", sheet_name='boardexcompanyid')
```

```python
boardex_df
```

```python
# Conver to list
boardex_list = boardex_df['Company Name'].tolist()
boardex_list = [x.lower() for x in boardex_list]
```

```python
missing_list = df['COMPANYNAME'].tolist()
missing_list = [x.lower() for x in missing_list]
```

```python
# Do fuzzy matching on the list of strings. 
# Compare the company name of each firm of interest against all the company names in the Boardex list, and return the top 2 matches.
firm_name_top_match_1 = []
firm_name_top_match_2 = []

for firm_name in missing_list:
    top_match = process.extract(firm_name.lower(), boardex_list, scorer=fuzz.WRatio, limit=2)
    print(firm_name, top_match[0], top_match[1])
    firm_name_top_match_1.append(top_match[0])
    firm_name_top_match_2.append(top_match[1])
```

```python
# Build the df with the string match results
zipped = list(zip(missing_list, firm_name_top_match_1, firm_name_top_match_2))
df_output = pd.DataFrame(zipped, columns=['missing_list', 'firm_name_top_match_1', 'firm_name_top_match_2'])

firm_name_top_match_1_score = []
firm_name_top_match_1_name = []
firm_name_top_match_1_index = []
for row in firm_name_top_match_1:
    name = row[0]
    score = row[1]
    index = row[2]
    firm_name_top_match_1_name.append(name)
    firm_name_top_match_1_score.append(score)
    firm_name_top_match_1_index.append(index)
    
firm_name_top_match_2_score = []
firm_name_top_match_2_name = []
firm_name_top_match_2_index = []
for row in firm_name_top_match_2:
    name = row[0]
    score = row[1]
    index = row[2]
    firm_name_top_match_2_name.append(name)
    firm_name_top_match_2_score.append(score)
    firm_name_top_match_2_index.append(index)

df_output['firm_name_top_match_1_name'] = firm_name_top_match_1_name
df_output['firm_name_top_match_1_score'] = firm_name_top_match_1_score
df_output['firm_name_top_match_1_index'] = firm_name_top_match_1_index
df_output['firm_name_top_match_2_name'] = firm_name_top_match_2_name
df_output['firm_name_top_match_2_score'] = firm_name_top_match_2_score
df_output['firm_name_top_match_2_index'] = firm_name_top_match_2_index
```

```python
df_output.to_excel(r"C:\Users\jasonjia\Dropbox\Projects\boardex_and_execucomp\string_similarity_check_for_missing_isins_full.xlsx", index=False)
```

# Part 3: Update original isin

```python
# Read in the spreadsheet with manual checking of strings done
df_withresult = pd.read_excel(r"C:\Users\jasonjia\Dropbox\Projects\boardex_and_execucomp\string_similarity_check_for_missing_isins_full_withresult.xlsx")
```

```python
df_withresult
```

```python
boardex_df['index'] = boardex_df.index
boardex_df
```

```python
# Get the matching index: -1 if no match, and the corresponding index if there's a match with the 
# 1st or 2nd Boardex company name.
index_list = []
for index, row in df_withresult.iterrows():
    result = row['result']
    if result == 0:
        index = -1
    elif result == 1:
        index = row['firm_name_top_match_1_index']
    else:
        index = row['firm_name_top_match_2_index']
    index_list.append(index)
df_withresult['index_to_match'] = index_list
```

```python
df_withresult
```

```python
# Merge company names with ISINs and gvkeys from list of interest
df_merge = df_withresult.merge(boardex_df, how='left', left_on='index_to_match', right_on='index')
df_merge
```

```python
# Get columns of interest
df_final = df_merge.loc[:, ['missing_list', 'ISIN', 'CompanyID*', 'Company Name', 'Country', 'Ticker', 'index']]
```
