# Nationality mix

```python
df_nationality = df_with_manager[['DirectorID*','Nationality']]
df_nationality = df_nationality.drop_duplicates(subset=['DirectorID*'], ignore_index=True)
df_nationality
```

```python
df_with_nationality = df.merge(df_nationality, on='DirectorID*')
df_with_nationality = df_with_nationality.drop_duplicates(subset=['CompanyID*','DirectorID*'], ignore_index=True)
df_with_nationality
```

```python
def most_common_nationalities(x):
    nationalities_value_counts = x.value_counts(dropna=True)
    return [most_common_nationality(nationalities_value_counts, 1),
            most_common_nationality(nationalities_value_counts, 2),
            most_common_nationality(nationalities_value_counts, 3)]

def most_common_nationality(nationalities_value_counts, rank):
    if len(nationalities_value_counts) < rank:
        return np.NaN
    else:
        return nationalities_value_counts.index[rank-1]
```

```python
agg_dict = {'Nationality Mix':lambda x:most_common_nationalities(x)}
# Note: CompanyID and CompanyID + Company Name both gave same number of rows
df_with_nationality_groupby_firm = df_with_nationality.groupby(['CompanyID*', 'Company Name']).agg(agg_dict).reset_index()
df_with_nationality_groupby_firm 
```

```python
df_with_nationality_groupby_firm['nationality_1'] = df_with_nationality_groupby_firm['Nationality Mix'].str[0]
df_with_nationality_groupby_firm['nationality_2'] = df_with_nationality_groupby_firm['Nationality Mix'].str[1]
df_with_nationality_groupby_firm['nationality_3'] = df_with_nationality_groupby_firm['Nationality Mix'].str[2]
df_with_nationality_groupby_firm
```
