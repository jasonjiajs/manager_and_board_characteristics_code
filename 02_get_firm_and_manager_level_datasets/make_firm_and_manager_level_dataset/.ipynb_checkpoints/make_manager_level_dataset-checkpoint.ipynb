## Get dataset of all directors

```python
# Check that directorid* and director name match 1:1. Yes.
df_sector1 = df.drop_duplicates(subset=['DirectorID*', 'Sector'], ignore_index=True)
df_sector1
```

```python
df_sector2 = df.drop_duplicates(subset=['DirectorID*', 'Individual Name', 'Sector'], ignore_index=True)
df_sector2
```

# Get whether a manager has worked in a 'finance' sector

```python
df_sector2['worked_in_finance'] = df_sector2['Sector'].isin(['Banks', 'Investment Companies', 'Private Equity'])
df_sector2
```

```python
directors_worked_in_finance = df_sector2[df_sector2['worked_in_finance'] == True]
directors_worked_in_finance = directors_worked_in_finance[['DirectorID*', 'Individual Name']].drop_duplicates(ignore_index=True)
directors_worked_in_finance
```

```python
directors_worked_in_finance_list = list(directors_worked_in_finance['DirectorID*'])
directors_worked_in_finance_list
```

```python
df_sector2['director_worked_in_finance'] = df_sector2['DirectorID*'].isin(directors_worked_in_finance_list)
df_sector2
```

```python
test = df_sector2[df_sector2['director_worked_in_finance']]
print(test['DirectorID*'].unique)
test[test['DirectorID*'] == 11221963228]
```

# Get df manager

```python
df_manager = df_sector2.drop_duplicates(subset=['DirectorID*', 'Individual Name'], ignore_index=True)
df_manager
```

```python
df_manager_final = df_manager[['DirectorID*', 'Individual Name', 'Individual Role', 'director_worked_in_finance', 'Director Type']]
df_manager_final
```

# Get ceo/cfo

```python
indv_role = list(df_manager_final['Individual Role'].unique())
print(indv_role)
```

```python
# All roles starting with ceo = 1 
# Chief Executive Officer only appears once in the list of all possible roles, as 
# 'Division Interim President/Chief Executive Officer', so it's excluded
r = re.compile("^ceo", flags=re.I)
ceo_roles = list(filter(r.match, indv_role))
print("ceo roles:", ceo_roles)

print("")

# All roles starting with cfo = 1
# Chief Financial Officer only appears once in the list of all possible roles, as
# 'Partner/Division Chief Financial Officer', so it's excluded
r = re.compile("^cfo", flags=re.I)
cfo_roles = list(filter(r.match, indv_role)) 
print("cfo roles:", cfo_roles)
```

```python
df_sector2['ceo'] = df_sector2['Individual Role'].isin(ceo_roles)
df_sector2['cfo'] = df_sector2['Individual Role'].isin(cfo_roles)
df_sector2
```

```python
# Get a list of directors who have been CEOs or CFOs
ceo_list = list(df_sector2[df_sector2['ceo']]['DirectorID*'].unique())
cfo_list = list(df_sector2[df_sector2['cfo']]['DirectorID*'].unique())
print('number of ceos:', len(ceo_list))
print('number of cfos:', len(cfo_list))
```

## Groupby Director ID

```python
agg_dict = {'Age':'max','Company Name': 'count', 'Director Type':'first','director_worked_in_finance':'max','ceo':'max','cfo':'max','Nationality':'first'}
df_merge_groupby = df_merge.groupby(['DirectorID*', 'Individual Name','DOB']).agg(agg_dict).reset_index()
df_merge_groupby
```

```python
df_merge_groupby['ceo'] = df_merge_groupby['DirectorID*'].isin(ceo_list)
df_merge_groupby['cfo'] = df_merge_groupby['DirectorID*'].isin(cfo_list)
df_merge_groupby
```

# Merge with Characteristics - Education Dataset

```python
df3 = pd.read_csv(r"C:\Users\jasonjia\Dropbox\Projects\boardex_and_execucomp\output\filter_boardex_by_cols\output_from_mercury_combined_fragment_country_and_managementtype\Director Profile - Education & Achievements - sheet0_director_and_smde.csv", low_memory=False)
df3
```
