```python
# Import Packages
import pandas as pd, numpy as np
from pathlib import Path
import os, sys, glob, re
```

# Define input and output files

```python
def parent_dir(dir):
    return os.path.abspath(os.path.join(dir, os.pardir))
```

```python
# Define input folder containing output from combined BoardEx files
inputfilepath = Path(parent_dir(parent_dir(os.getcwd())) / Path('output/02_get_initial_datasets/02.2_boardex_combined/Board Summary.csv'))
print('inputfilepath:', inputfilepath, inputfilepath.__class__)
# Define output folder containing firm-level dataset
outputfilepath = Path(parent_dir(parent_dir(os.getcwd())) / Path('output/03_get_firm_and_manager_level_datasets/board_summary_isins.txt'))
print('outputfilepath:', outputfilepath, outputfilepath.__class__)
```

# Import Board Summary

```python
df = pd.read_csv(inputfilepath, low_memory=False)
df
```

# Get list of unique ISINs, company IDs and company names

```python
print('Number of unique (ISIN) combinations:', df[['ISIN']].drop_duplicates().shape[0])
print('Number of unique (ISIN, Company ID) combinations:', df[['ISIN', 'CompanyID*']].drop_duplicates().shape[0])
print('Number of unique (ISIN, Company Name) combinations:', df[['ISIN', 'Company Name']].drop_duplicates().shape[0])
print('Number of unique (Company ID) combinations:', df[['CompanyID*']].drop_duplicates().shape[0])
print('Number of unique (Company ID, Company Name) combinations:', df[['CompanyID*', 'Company Name']].drop_duplicates().shape[0])
print('Number of unique (Company Name) combinations:', df[['Company Name']].drop_duplicates().shape[0])
print('Number of unique (ISIN, Company ID, Company Name) combinations:', df[['ISIN', 'CompanyID*', 'Company Name']].drop_duplicates().shape[0])
```

```python
# View df of unique (ISIN, Company ID, Company Name) combinations
df[['CompanyID*', 'ISIN', 'Company Name']].drop_duplicates()
```

```python
# Get df of unique ISINs
df_output = df[['ISIN']].drop_duplicates()

# Some rows have two or more ISINs separated by ','. Take only the first ISIN using regex.
df_output['ISIN'] = df_output['ISIN'].str.split(',').str[0]

# Save as txt file (to be fed into WRDS later on)
df_output.to_csv(outputfilepath, index=False, header=False)
df_output
```

# Next steps: Map firm identifier (e.g. ISIN) in BoardEx to gvkeys in Compustat
- Initially we wanted to map Company IDs to gvkeys, but we don't have access to the [BoardEx CRSP Compustat Link (Beta)](https://wrds-www.wharton.upenn.edu/pages/get-data/linking-suite-wrds/boardex-crsp-compustat-link/).
- Instead, we use the [Compustat Global - Fundamentals Annual](https://wrds-www.wharton.upenn.edu/pages/get-data/compustat-capital-iq-standard-poors/compustat/global-daily/fundamentals-annual/) crosswalk to map ISINs to gvkeys.
- While there's no option to select gvkey as a desired output column, it will automatically be included in the output.

```python
# Read in ISIN to gvkey crosswalk
isin_gvkey_crosswalk_filepath = Path(parent_dir(parent_dir(os.getcwd())) / Path('output/03_get_firm_and_manager_level_datasets/board_summary_isin_gvkey_crosswalk.csv'))
isin_gvkey_crosswalk = pd.read_csv(isin_gvkey_crosswalk_filepath)
isin_gvkey_crosswalk
```

```python
# Get unique (ISIN, gvkey) combinations
isin_gvkey_crosswalk = isin_gvkey_crosswalk[['gvkey', 'isin']].drop_duplicates()
isin_gvkey_crosswalk
```

# Merge gvkeys to firm-level dataset on ISINs

```python
# Merge this back to Board Summary
df_firm_filepath = Path(parent_dir(parent_dir(os.getcwd())) / Path('output/03_get_firm_and_manager_level_datasets/boardex_firm_level_dataset.csv'))
df_firm = pd.read_csv(df_firm_filepath)

# Get first ISIN for each firm
df_firm['isin'] = df_firm['isin'].str.split(',').str[0]

# Merge with gvkeys
df_firm_gvkey = df_firm.merge(isin_gvkey_crosswalk, on='isin', how='left')

# Save df_firm_gvkey
df_firm_gvkey_filepath = Path(parent_dir(parent_dir(os.getcwd())) / Path('output/03_get_firm_and_manager_level_datasets/boardex_firm_level_dataset_gvkey.csv'))
df_firm_gvkey.to_csv(df_firm_gvkey_filepath, index=False)
df_firm_gvkey_filepath = Path(parent_dir(parent_dir(os.getcwd())) / Path('output/03_get_firm_and_manager_level_datasets/boardex_firm_level_dataset_gvkey.csv'))
df_firm_gvkey.to_stata(df_firm_gvkey_filepath, write_index=False, version=117)

df_firm_gvkey
```
